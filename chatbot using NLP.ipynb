```python
import spacy
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
# Load Spacy model
nlp = spacy.load("en_core_web_sm")

# Example text
text = "we are good"
doc = nlp(text)

# Tokenize and stem
tokens = [token.text for token in doc]
ps = PorterStemmer()
stemming = [ps.stem(word) for word in tokens]
processed_text = ' '.join(stemming)

# Prepare the corpus for TfidfVectorizer
corpus = [processed_text]

# Replies dictionary
replies = {
    "how are you": "I am fine",
    "hello why are you happy": "how i can help you"
}

# Process replies and add to corpus
for i in replies.keys():
    stemm = [ps.stem(word) for word in i.split()]
    processed_text = ' '.join(stemm)
    corpus.append(processed_text)

# Fit the TfidfVectorizer on the entire corpus
tfidf = TfidfVectorizer()
tfidf.fit(corpus)

# Transform the initial text
vectors = tfidf.transform([corpus[0]]).toarray()
print(f"Vector for the initial text: {vectors}")

# Transform and print vectors for the replies
for i, processed_text in zip(replies.keys(), corpus[1:]):
    vectors = tfidf.transform([processed_text]).toarray()
    print(f"Vector for '{i}': {vectors}")
    print(f"Stemmed: {processed_text.split()}")
```

# Using Machine learning NLP nltk + spacy + transformers + Sklearn

```python
import spacy
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load Spacy model
nlp = spacy.load("en_core_web_sm")

# Tokenize and stem function
def tokenize_and_stem(text, nlp, ps):
    doc = nlp(text)
    tokens = [token.text for token in doc]
    stemming = [ps.stem(word) for word in tokens]
    return ' '.join(stemming)


# Initialize PorterStemmer
ps = PorterStemmer()
text=input()
corpus = [tokenize_and_stem(text, nlp, ps)]

# Replies dictionary
replies = {
    "how are you": "I am fine",
    "hello why are you happy": "how i can help you",
    "what is your name": "I am a evochat",
    "what do you do": "I chat with users",
    "how is the weather": "I do not have access to weather information",
    "tell me a joke": "Why don't scientists trust atoms? Because they make up everything!",
    "what is your favorite color": "I like all colors equally",
    "how old are you": "I am timeless",
    "where are you from": "I exist in the digital world",
    "what is your purpose": "To assist and provide information",
    "who created you": "I was created by evotechnology.us",
    "can you help me": "Of course! What do you need help with?",
    "tell me a fact": "Did you know? Honey never spoils",
    "what is the capital of France": "The capital of France is Paris",
    "what is the capital of Pak": "The capital of Pakistan is Islamabad",
    "do you like music": "I enjoy all kinds of music",
    "can you dance": "I can imagine dancing, but I cannot physically dance",
    "how can I learn programming": "There are many online resources and courses to learn programming",
    "what is AI": "AI stands for Artificial Intelligence",
    "do you have a family": "I am a virtual entity and do not have a family",
    "what are you": "I am a conversational AI chatbot",
    # Add more question-reply pairs as needed
}


# Process replies and add to corpus
for i in replies.keys():
    processed_text = tokenize_and_stem(i, nlp, ps)
    corpus.append(processed_text)
# Fit the TfidfVectorizer on the entire corpus
tfidf = TfidfVectorizer()
tfidf.fit(corpus)

# Transform the initial text and replies
vector_text = tfidf.transform([corpus[0]]).toarray()
vectors_replies = tfidf.transform(corpus[1:]).toarray()

# Calculate cosine similarities
cosine_similarities = cosine_similarity(vector_text, vectors_replies)

# Find the index of the most similar reply
most_similar_index = cosine_similarities.argmax()
print(most_similar_index)
# Get the corresponding reply
most_similar_reply_key = list(replies.keys())[most_similar_index]
reply = replies[most_similar_reply_key]

print(f"Input text: '{text}'")
print(f"Most similar reply: '{reply}'")
```

# Using Deep Learning NLP Tensorflow

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Sample data
data = {
    "hi": "Hello!",
    "how are you": "I am fine, thank you!",
    "what is your name": "I am a chatbot.",
    "what do you do": "I chat with users.",
    "tell me a joke": "Why don't scientists trust atoms? Because they make up everything!",
    "what is the capital of Pak": "The capital of Pakistan is Islamabad"
}

# Prepare the data
input_texts = []
target_texts = []
for question, answer in data.items():
    input_texts.append(question)
    target_texts.append('<start> ' + answer + ' <end>')

# Tokenize the data
tokenizer = Tokenizer(filters='')
tokenizer.fit_on_texts(input_texts + target_texts)
input_sequences = tokenizer.texts_to_sequences(input_texts)
target_sequences = tokenizer.texts_to_sequences(target_texts)

# Pad the sequences
max_input_length = max([len(seq) for seq in input_sequences])
max_target_length = max([len(seq) for seq in target_sequences])
encoder_input_data = pad_sequences(input_sequences, maxlen=max_input_length, padding='post')
decoder_input_data = pad_sequences(target_sequences, maxlen=max_target_length, padding='post')

# Create decoder target data
vocab_size = len(tokenizer.word_index) + 1
decoder_target_data = np.zeros((len(target_sequences), max_target_length, vocab_size))
for i, seq in enumerate(target_sequences):
    for t, word_id in enumerate(seq):
        if t > 0:
            decoder_target_data[i, t - 1, word_id] = 1.0

# Build the model
latent_dim = 256

# Encoder
encoder_inputs = Input(shape=(None,))
encoder_embedding = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=latent_dim)(encoder_inputs)
encoder_lstm = LSTM(latent_dim, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(encoder_embedding)
encoder_states = [state_h, state_c]

# Decoder
decoder_inputs = Input(shape=(None,))
decoder_embedding = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=latent_dim)(decoder_inputs)
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=encoder_states)
decoder_dense = Dense(vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

# Define the model
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy')

# Train the model
model.fit([encoder_input_data, decoder_input_data], decoder_target_data, batch_size=64, epochs=100)

# Inference setup
encoder_model = Model(encoder_inputs, encoder_states)

decoder_state_input_h = Input(shape=(latent_dim,))
decoder_state_input_c = Input(shape=(latent_dim,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

decoder_outputs, state_h, state_c = decoder_lstm(
    decoder_embedding, initial_state=decoder_states_inputs)
decoder_states = [state_h, state_c]
decoder_outputs = decoder_dense(decoder_outputs)

decoder_model = Model(
    [decoder_inputs] + decoder_states_inputs,
    [decoder_outputs] + decoder_states)

# Decode the sequence
def decode_sequence(input_seq):
    states_value = encoder_model.predict(input_seq)

    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = tokenizer.word_index['<start>']

    stop_condition = False
    decoded_sentence = ''
    while not stop_condition:
        output_tokens, h, c = decoder_model.predict(
            [target_seq] + states_value)

        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_word = tokenizer.index_word[sampled_token_index]
        decoded_sentence += ' ' + sampled_word

        if sampled_word == '<end>' or len(decoded_sentence) > max_target_length:
            stop_condition = True

        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_index

        states_value = [h, c]

    return decoded_sentence.strip()

# Test the model
def chat(input_text):
    input_seq = tokenizer.texts_to_sequences([input_text])
    input_seq = pad_sequences(input_seq, maxlen=max_input_length, padding='post')
    decoded_sentence = decode_sequence(input_seq)
    print('Chatbot:', decoded_sentence)

# Example conversation
text=input()
chat(text)
```
